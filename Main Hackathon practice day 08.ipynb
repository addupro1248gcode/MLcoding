Steps for NLP:-


1. Cleaning the text, by removing anything except numbers and alphabets.
2. Clean every extra space
Stemming- convert words to base word and avoid prurals, etc.
  Eg.

*   going-----> go,
*   doing-----> do,
*   timing----> time

      
      
4. Remove stop words - a, am, and, I, a

Library for cleaning is nltk(natural language tool kit)

```python
import pandas as pd
dataset=pd.read_csv('/content/Restaurant_Reviews.tsv',delimiter='\t')
import re # for cleaning of garbage values like- @,-,_,^,%
import nltk # for cleaning including spaces, split, stemma and lemma(stemma and lemma make word into base word)
from nltk.corpus import stopwords# list of stopwords
from nltk.stem import WordNetLemmatizer# converts words into base words
nltk.download('stopwords')
nltk.download('wordnet')

corpus = [] #Empty list for storing our word corpus(group or bag of words)
text_data = ""#
#pattern=re.sub('[^a-zA-Z0-9]')
for i in range(0,1000):
  text_data = re.sub('[^a-zA-Z0-9]', ' ', dataset['Review'][i])#re(regular expression) extracts patterns-[^a-zA-Z] only allows words and "Review" checks all the things in reviews column
  text_data = text_data.lower() #lowering the case of Alphabet
  text_data = text_data.split() # splitting the words for getting clear sentence
  wl = WordNetLemmatizer() # converting the word to its base word
  text_data = [wl.lemmatize(word) for word in text_data if not word in set(stopwords.words('english'))]
  text_data = ' '.join(text_data) # it removes the extra space after stemming
  corpus.append(text_data)

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1500)
X=cv.fit_transform(corpus).toarray()
y=dataset[['Liked']].values
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.20,random_state=42)
X_train.shape
y_train.shape
#Naive Base:This algo will work for NLP task with long sentence sequence
#It is classification algo
#For regression(RandomForestRegressor, Linear reg, poly nomial reg)
#if lables in data for classification(RandomForestClassifier, Decisiontree)
#NLp with sentences(naive_bayes,deepleatning)
#xgboost- will focus on wrong results and makes them right

#from sklearn.naive_bayes import GaussianNB
#classifier=GaussianNB()

from sklearn.ensemble import GradientBoostingClassifier
classifier=GradientBoostingClassifier()
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
print(y_pred)
#try this with xgboost, or GradientBoostingClassifier, original was GaussianNB() of naive_bayes

#to verify model results we use confusion matrix

import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import accuracy_score
confusion_matrix = metrics.confusion_matrix(y_test,y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
accuracy_score(y_test,y_pred)#for presenting accuracy score
#Scope of improvement
#-in data cleaning
#-model parameter tuning or model selection
```
